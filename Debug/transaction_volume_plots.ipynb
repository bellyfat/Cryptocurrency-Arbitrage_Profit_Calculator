```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
import os
```

```python
def plot_df_trade_volumes(arr, plot_title, col):
    df=pd.DataFrame(arr)
    df=df[df[6]==col]
    df[4]=df[4].apply(lambda x: datetime.datetime.utcfromtimestamp(x))
    df[4]=df[4].apply(lambda x: x.replace(second=0))
    d=df.groupby(4).agg({3:'sum'}).reset_index()
    if col==3:
        d[3]=d[3]*-1
    d[4]=pd.to_datetime(d[4])
    d=d.sort_values(4,ascending=True)
    plt.plot(d[4],d[3])
    plt.rcParams['figure.figsize'] = [10,10]
    plt.title(plot_title)
```

```python
arr1=np.load('update_order_ethusd_23.npy')
arr2=np.load('update_order_btcusd_23.npy')
arr3=np.load('update_order_ethbtc_23.npy')
```

```python
plot_df_trade_volumes(arr1,"eth to usd transactions",3)
```
